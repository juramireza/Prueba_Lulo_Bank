#Construcción Sabana inicial

#1.Definir listas para guardar errores

```python
log_error_list = []
```

#2.Importar librerías

```python
import requests
import pprint
import pandas as pd
import json
from datetime import date
import itertools
import glob
import csv
```

#3.Funciones

```python
#Convertir una lista de listas en una lista 
def aplanando_lista_de_lista(data):
    ##aplanando de la lista
    flat_list_2 = []
    for item in data:
      # hacer un append de los elementos de flat_list
      flat_list_2 += item

    df=flat_list_2
    return df
```

#4.Lectura de los json y cosolidación de un dataframe

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
      #Se guardan las rutas de los archivos json en json_files
      json_files = glob.glob('/dbfs/mnt/raw/json/*.json')

      #Se leen los archivos json
      list_data = []
        
      # Escribimos un for que irá a través de cada uno de los nombres de archivo a través de globbing y el resultado final será la lista de achivos json
      for filename in json_files:
        with open(filename) as archivo: datos=json.load(archivo)
        list_data.append(datos)

      #Se aplana la lista de resultados 
      lista_de_resultados=aplanando_lista_de_lista(list_data) 

      # normalizacion del json a dataframe
      datos = pd.json_normalize(lista_de_resultados) 
      
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```

#5.Escribir la sabana inicial en Bronze

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
      datos.to_csv("/dbfs/mnt/processed/Sabana_inicial/Sabana_inicial.csv",index=False)
      
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```
