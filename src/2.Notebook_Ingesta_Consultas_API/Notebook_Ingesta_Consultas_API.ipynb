#2.Importar librerías

```python
import requests
import pprint
import pandas as pd
import json
from datetime import date
import itertools
```

#3.Funciones para crear dataframe de fechas y para aplanar una lista

```python
# Función para crear una tabla de fechas dado una fecha inicial y fecha final:

def create_date_table(start='2020-12-01', end='2020-12-31'):
    df = pd.DataFrame({"Date": pd.date_range(start, end)})
    return df

# Función aplanar una lista (convertir una lista de listas en una lista):

def aplanando_lista_de_lista(data):
    ##aplanando de la lista
    flat_list_2 = []
    for item in data:
      # hacer un append de los elementos de flat_list
      flat_list_2 += item

    df=flat_list_2
    return df
```

#4.Se crea la lista de fechas con base en los parámetros dados

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
        #Se crea la tabla de fechas CON la funcion create_date_table()
        fechas = create_date_table() 

        #Se convierte la columna de fechas a texto para más adelante concatenarla con la url
        fechas['Date']=fechas['Date'].astype(str)

        #Se convierte el dataframe a lista
        fechas = fechas.to_numpy().tolist() 

        # Se emplea la funcion para aplanar la lista de lista de fechas
        lista_fechas=aplanando_lista_de_lista(fechas) 
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```

#5.Extracción e ingesta de las consultas del API json

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
      lista = []
      for i in lista_fechas:
          #Peticiones de la url por medio del for
          url = "http://api.tvmaze.com/schedule/web?date={}".format(i) 
          #Petición get a la url
          response = requests.get(url)
          #Se convierte la cadena de texto a diccionario 
          json_dict= json.loads(response.content)

          #Se escribe cada uno de los archivos json
          with open('/dbfs/mnt/raw/json/Date_{}.json'.format(i), 'w') as file:
            json.dump(json_dict, file)
          
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```
