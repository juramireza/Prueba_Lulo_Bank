#Transformaciones sobre el dataframe embedded_show_net

#1.Definir listas para guardar errores

```python
log_error_list = []
```

#2.Instalación de paquetes e importación de librerías

```python
#Se importan las librerías a utilizar 
import requests
import pprint
import pandas as pd
import json
import glob
import sys
from datetime import date, time, datetime
import sqlite3
```

#3.Lectura de la sabana de datos inicial

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
      ##Se lee la sabana inicial de datos 
      Sabana_inicial = pd.read_csv("/dbfs/mnt/processed/Sabana_inicial/Sabana_inicial.csv", sep=',')
      
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```

#4.Dataframe de embedded_show por el _embedded.show.network.id

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
        #Se seleccionan las columnas para formar el dataframe de embedded_show_net_work
        embedded_show_net_work = Sabana_inicial[[  '_embedded.show.network.id',
                                                   '_embedded.show.network.name',
                                                   '_embedded.show.network.country.name',
                                                   '_embedded.show.network.country.code',
                                                   '_embedded.show.network.country.timezone',
                                                   '_embedded.show.network.officialSite']]

        #Filtrar los registros que no tengan _embedded.show.id
        embedded_show_net_work  = embedded_show_net_work[embedded_show_net_work['_embedded.show.network.id'].notnull()]
        
        
        #Eliminar duplicado por el _embedded.show.id
        embedded_show_net_work = embedded_show_net_work.drop_duplicates(subset=['_embedded.show.network.id'])
        
        #Renombrar columnas 
        embedded_show_net_work = embedded_show_net_work.rename(columns={'_embedded.show.network.id':'embedded_show_network_id',
                                                                              '_embedded.show.network.name':'embedded_show_network_name',
                                                                              '_embedded.show.network.country.name':'embedded_show_network_country_name',
                                                                              '_embedded.show.network.country.code':'embedded_show_network_country_code',
                                                                              '_embedded.show.network.country.code':'embedded_show_network_country_cod',
                                                                              '_embedded.show.network.country.timezone':'embedded_show_network_country_timezone',
                                                                              '_embedded.show.network.officialSite':'embedded_show_network_officialSite'})
        #Reiniciar el índice
        embedded_show_net_work.reset_index(level=None, drop=True, inplace=True, col_level=0, col_fill='')
      
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```

#5.Escritura del dataframe en processed

```python
if len(log_error_list)>0:
    print("Previous Error")
else:
    try:
        embedded_show_net_work.to_csv("/dbfs/mnt/processed/Embedded_show_net_work/embedded_show_net_work.csv",index=False)
      
    except Exception as e:
        print("Failed! Due to: {}".format(str(e)))
        log_error_list.append("Error: {}".format(str(e)))
```
