###1.Importar librerías

```
import pyspark
from delta.tables import *
from pyspark.sql.types import *
from pyspark.sql.types import *
```

###2.Definición del esquema de la tabla Dim_Tiempo

```
#Se define el esquema de la tabla de Dim_Tiempo
Dim_Tiempo_schema=StructType(fields=[StructField("Date",TimestampType(),False),
                                   StructField("Day",LongType(),True),
                                   StructField("DayName",StringType(),True),
                                   StructField("WeekDayNum",LongType(),True),
                                   StructField("Quarter",LongType(),True),
                                   StructField("Year",LongType(),True),
                                   StructField("Year_half",LongType(),True),
                                   StructField("MonthNum",LongType(),True),
                                   StructField("MonthLong",StringType(),True),
                                   StructField("Month",StringType(),True),
                                   StructField("MonthYear",StringType(),True),
                                   StructField("MonthYearLong",StringType(),True)          
                                  ])
#Se crea una Delta Table
Dim_Tiempo= sqlContext.createDataFrame([],Dim_Tiempo_schema)#Create Delta Table

#Se escribe la Delta Table de Dim_Tiempo
Dim_Tiempo.write.mode("overwrite").format("delta").save("/mnt/presentation/Dim_Tiempo")
```
