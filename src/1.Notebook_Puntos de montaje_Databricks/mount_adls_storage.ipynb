##3.Puntos de montaje

###3.1 Primer punto de montaje

```
dbutils.fs.mount(
  source = "wasbs://"+blob_storage_container1+"@"+blob_storage_account+".blob.core.windows.net",
  mount_point = "/mnt/raw",
  extra_configs = {'fs.azure.account.key.' + blob_storage_account + '.blob.core.windows.net': blob_storage_key})
```

###3.1 Segundo punto de montaje

```
dbutils.fs.mount(
  source = "wasbs://"+blob_storage_container2+"@"+blob_storage_account+".blob.core.windows.net",
  mount_point = "/mnt/processed",
  extra_configs = {'fs.azure.account.key.' + blob_storage_account + '.blob.core.windows.net': blob_storage_key})
```
